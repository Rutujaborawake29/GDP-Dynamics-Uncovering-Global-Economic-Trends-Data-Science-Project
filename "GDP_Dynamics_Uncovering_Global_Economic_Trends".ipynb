Group by "Country Name" and find the average GDP per capita across all years.


```python
avg_gdp_per_year=df.iloc[:, 4:].mean(axis=1)
avg_gdp_per_year
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
avg_gdp_per_year.plot(kind="line", marker="o", color="b", linestyle="-")
plt.title("Average GDP Per Capita Over the Years")
plt.xlabel("Year")
plt.ylabel("GDP Per Capita (Mean)")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.barplot(x=["2010"], y= df["2010"].sum(), palette="Blues")
plt.title("Total GDP Per Capita for All Countries in 2010")
plt.ylabel("Total GDP Per Capita")
plt.show()
```

```python
df["Average GDP Per Capita"] = df.iloc[:, 4:].mean(axis=1)
df.columns
```

```python
top_10_countries = df[["Country Name", "Average GDP Per Capita"]].sort_values(by="Average GDP Per Capita", ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x="Average GDP Per Capita", y="Country Name", data=top_10_countries, palette="magma")
plt.title("Top 10 Countries by Average GDP Per Capita")
plt.xlabel("Average GDP Per Capita")
plt.ylabel("Country Code")
plt.show()
```
